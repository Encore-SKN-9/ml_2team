```python
import pandas as pd
```

```python
df = pd.read_csv('./Data/preprocessed_movies_data.csv')

df.head()
```

```python
# 콜론으로 구분된 데이터의 첫번째 데이터만 추출
def extractFirstStr(input):
    if type(input) == str:
        return input.split(',')[0]
    else: # 문자열 타입이 아닌 경우 그대로 반환
        return input 

# 감독, 배우, 작가, 제작사 데이터 중 첫번째 데이터만 추출
df['directors'] = df['directors'].apply(extractFirstStr)
df['stars'] = df['stars'].apply(extractFirstStr)
df['writers'] = df['writers'].apply(extractFirstStr)
df['production_companies'] = df['production_companies'].apply(extractFirstStr)

# 촬영지, 언어, 국가 데이터는 사용하지 않으므로 삭제
df = df.drop(['countries_origin', 'filming_locations', 'Languages'], axis=1)
# 수상 정보는 사용하지 않으므로 삭제
df = df.drop(['nominations','oscars'], axis=1)
# 평점 정보 삭제
df = df.drop(['Votes','Rating'], axis=1)
# 장르 데이터는 카테고리로 대체하므로 삭제
df = df.drop(['genres'], axis=1)

df.head()
```

```python
# 수익률 계산
df['Profit'] = df['grossWorldWide'] / df['budget'] * 100 
# 수익률이 100% 이상이면 흥행작으로 판단
df['Hit'] = df['Profit'] > 100

# 수익 관련된 컬럼 삭제
df = df.drop(['grossWorldWide', 'gross_US_Canada', 'opening_weekend_Gross'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# 레이블 인코더 생성

le = LabelEncoder()

# 각 컬럼에 대해 레이블 인코딩 수행
df['directors'] = le.fit_transform(df['directors'])
df['writers'] = le.fit_transform(df['writers'])
df['stars'] = le.fit_transform(df['stars'])
df['production_companies'] = le.fit_transform(df['production_companies'])
df['Category'] = le.fit_transform(df['Category'])
df['Hit'] = le.fit_transform(df['Hit'])

df.head()
```

```python
df.to_csv('./Data/encoded_movies_data.csv', index=False)
```
